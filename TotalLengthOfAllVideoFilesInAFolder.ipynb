```python
import moviepy.editor

# Converts into more readable format
def convert(seconds):
    hours = seconds // 3600
    seconds %= 3600

    mins = seconds // 60
    seconds %= 60

    return hours, mins, seconds

def SingleVideoDuration(Video_Directory):
  # Create an object by passing the location as a string
  video = moviepy.editor.VideoFileClip(Video_Directory)

  # Contains the duration of the video in terms of seconds
  video_duration = int(video.duration)
  
  return video_duration
```

```python
import os
from fnmatch import fnmatch

root = r'F:\Java Programming Masterclass for Software Developers'
pattern = "*.mp4"

video_duration = 0

i = 0 
for path, subdirs, files in os.walk(root):
    for name in files:
        if fnmatch(name, pattern):
            Video_Directory = os.path.join(path, name)
            i = i + 1
            print()
            print("Number Of Files: ", i)
            print("Video Directory Name: ", Video_Directory)
            video_duration = SingleVideoDuration(Video_Directory) + video_duration
            hours, mins, secs = convert(video_duration)
            print("Total Duration Calculation: ", hours, "hours", mins, "minutes", secs, "seconds")
```
